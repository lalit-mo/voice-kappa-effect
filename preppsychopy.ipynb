```python
from pydub import AudioSegment
import os
```

```python
def raise_pitch_octave(audio_path, output_path):
    # Load the audio file
    sound = AudioSegment.from_file(audio_path)
    
    # Extract raw audio data as NumPy array
    samples = np.array(sound.get_array_of_samples())
    
    # Define the pitch shift factor for raising by an octave
    pitch_shift_factor = 2.0 ** (1/12)  # Raising by an octave corresponds to a pitch shift of 12 semitones
    
    # Perform pitch shift by manipulating the raw audio samples
    shifted_samples = np.interp(
        np.arange(0, len(samples), pitch_shift_factor),
        np.arange(0, len(samples)),
        samples
    ).astype(np.int16)
    
    # Create a new AudioSegment from the shifted samples
    sound_octave_up = AudioSegment(
        shifted_samples.tobytes(),
        frame_rate=sound.frame_rate,
        sample_width=sound.sample_width,
        channels=sound.channels
    )
    
    # Export the modified audio
    sound_octave_up.export(output_path, format="wav")
    return output_path
```

```python
def convert_sample_rate(input_path, output_path, target_sample_rate=48000):
    # Load the audio file
    sound = AudioSegment.from_file(input_path)

    # Resample to the target sample rate
    resampled_sound = sound.set_frame_rate(target_sample_rate)

    # Export the resampled audio
    resampled_sound.export(output_path, format="wav")
```

```python
def counter_balance(df, available_sets_):
    #merge to create only two columns
    dfk = pd.DataFrame(columns=['word', 'path'])
    for i in range(1, 5):
        tempdf = pd.DataFrame(columns=['word', 'path'])
        tempdf['word'] = df[f'word{i}'].values
        tempdf['path'] = df[f'path{i}'].values
        dfk = pd.concat([dfk,tempdf])
    #removing words that start with n because \n is psychopy is next line
    dfk = dfk[~dfk['word'].str.startswith('n')]

    #sample_df from rhyming words
    sampled_df = df.loc[list(available_sets_)].sample(4)
    #update available_sets for next df
    available_sets_ = available_sets_ - set(sampled_df.index)

    #counterbalancing
    for _ in range (4):
        k = dfk.sample(4)
        new_row = pd.DataFrame([np.concatenate((k['path'].values, k['word'].values), axis=0)], columns=df.columns)
        sampled_df = pd.concat([sampled_df, new_row], ignore_index=True)
    sampled_df.sample(frac=1).reset_index(drop=True)
    sampled_df['delay'] = [0, 200, 400, 500, 550, 600, 800, 1000]
    return sampled_df, available_sets_
```
